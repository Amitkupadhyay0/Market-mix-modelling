Frequentist vs Bayesian

<hr style="border:0; border-top:1px solid #BBD7F2;">

<h3 style="color:#1565C0; font-weight:600;">1. Meaning of Probability</h3>
<p style="color:#333;">
<b style="color:#0D47A1;">Frequentist:</b> Probability is the <i>long-run frequency</i> of events. <br>
<b style="color:#C2185B;">Bayesian:</b> Probability expresses <i>belief</i> about uncertainty.
</p>

<p style="color:#333; margin-left:10px;">
<b>Example — Probability of rain</b><br>
<span style="color:#0D47A1;">Frequentist:</span> “It rains on 30% of similar past days.”<br>
<span style="color:#C2185B;">Bayesian:</span> “Given belief and information, chance of rain tomorrow is 30%.”
</p>

<hr style="border:0; border-top:1px solid #BBD7F2;">

<h3 style="color:#1565C0; font-weight:600;">2. Treatment of θ (unknown parameter)</h3>
<p style="color:#333;">
<b style="color:#0D47A1;">Frequentist:</b> θ is <b>fixed but unknown</b>. Data is random.<br>
<b style="color:#C2185B;">Bayesian:</b> θ is <b>random</b> with a distribution. Data becomes fixed once observed.
</p>

<p style="color:#333; margin-left:10px;">
<b>Example (coin with 7H, 3T):</b><br>
<span style="color:#0D47A1;">Frequentist:</span> θ̂ = 0.7<br>
<span style="color:#C2185B;">Bayesian:</span> θ ~ Beta(8, 4)
</p>

<hr style="border:0; border-top:1px solid #BBD7F2;">

<h3 style="color:#1565C0; font-weight:600;">3. Use of Prior Knowledge</h3>
<p style="color:#333;">
<b style="color:#0D47A1;">Frequentist:</b> No prior knowledge used — depends only on data.<br>
<b style="color:#C2185B;">Bayesian:</b> Naturally uses prior information + updates with data.
</p>

<p style="color:#333; margin-left:10px;">
<b>Example:</b> If coin is known to be near-fair, Bayesian uses this. Frequentist cannot.
</p>

<hr style="border:0; border-top:1px solid #BBD7F2;">

<h3 style="color:#1565C0; font-weight:600;">4. Interpretation of Intervals</h3>
<p style="color:#333;">
<b style="color:#0D47A1;">Frequentist Confidence Interval:</b>  
“Over many repeated experiments, 95% of intervals contain θ.”  
<br>
<b style="color:#C2185B;">Bayesian Credible Interval:</b>  
“Given data, θ has 95% probability of lying in this interval.”
</p>

<p style="color:#333; margin-left:10px;">
<b>Example:</b> CI = [0.55, 0.80] ⇒ θ is 95% likely inside this range (Bayesian view).
</p>

<hr style="border:0; border-top:1px solid #BBD7F2;">

<h3 style="color:#1565C0; font-weight:600;">5. Output Form</h3>
<p style="color:#333;">
<b style="color:#0D47A1;">Frequentist:</b> Produces a <b>single estimate</b>, e.g., θ̂ = 0.7.  
<br>
<b style="color:#C2185B;">Bayesian:</b> Produces a <b>full distribution</b> for θ, showing uncertainty.
</p>

<p style="color:#333; margin-left:10px;">
<b>Example:</b> Posterior Beta(8,4) shows confidence/spread around θ.
</p>

###### Below is an Example to understand Bayesian vs Frequency estimation

<div style="border:2px solid #6A5ACD; padding:22px; border-radius:14px; background:#F9F9FF; font-family:Arial, sans-serif;">

### <span style="color:#5A3EC8;">Bayesian Estimation: Coin-Toss Example</span>

To demonstrate **Bayesian thinking**, we use a small dataset of coin tosses:

**Data:** <span style="color:#FF4500;">{ H, H, T, H, T, H, H, H, T, H }</span>

- <span style="color:#228B22;">Heads</span> = 7  
- <span style="color:#B22222;">Tails</span> = 3  

Our goal is to estimate the parameter:

$$
\theta = \text{Probability of getting Heads}
$$

---

### <span style="color:#5A3EC8;">1. Prior Distribution (Belief Before Data)</span>

We choose a Beta$(a, b)$ prior. For simplicity, assume a **uniform prior**:

$$
\theta \sim \text{Beta}(1,1)
$$

This represents:

- No strong belief  
- All values $\theta \in [0,1]$ equally likely  

**Reason for choosing Beta:**  
<span style="color:#8A2BE2;">Beta</span> is the conjugate prior of the **Binomial likelihood**, which means the posterior will also be Beta, making the update simple and analytical.

---

### <span style="color:#5A3EC8;">2. Likelihood From Data</span>

For coin tosses, the likelihood is **Binomial**:

$$
p(x|\theta) = \theta^{H}(1-\theta)^{T}
$$

Here:

$$
H = 7, \quad T = 3
$$

$$
p(x|\theta) = \theta^{7}(1-\theta)^{3}
$$

---

### <span style="color:#5A3EC8;">3. Posterior Distribution (Updated Belief)</span>

Bayes rule:

$$
p(\theta|x) \propto p(x|\theta) \cdot p(\theta)
$$

Substituting the prior and likelihood:

$$
p(\theta|x) \propto \theta^{7}(1-\theta)^{3} \cdot \theta^{1-1}(1-\theta)^{1-1}
$$

Simplifying:

$$
p(\theta|x) \propto \theta^{7}(1-\theta)^{3}
$$

This has the same functional form as a **Beta distribution**:

$$
\theta | x \sim \text{Beta}(1+7, 1+3)
$$

$$
\theta | x \sim \text{Beta}(8, 4)
$$

---

### <span style="color:#5A3EC8;">4. Posterior Mean (Bayesian Estimate)</span>

Posterior mean of Beta$(a, b)$:

$$
E[\theta|x] = \frac{a}{a+b}
$$

For Beta(8,4):

$$
E[\theta|x] = \frac{8}{8+4} = 0.6667
$$

Thus the **Bayesian estimate**:

$$
\hat{\theta}_{\text{Bayesian}} = 0.667
$$

The posterior distribution also gives **uncertainty**, not just a single number.

---

### <span style="color:#5A3EC8;">5. Final Interpretation</span>

The data gave <span style="color:#228B22;">7 heads</span> in 10 tosses, but the prior softened the estimate from the **frequentist estimate 0.70** down to <span style="color:#8B008B;">0.667</span>, reflecting mild prior belief that all $\theta$ were equally likely.  

This combination of:

$$
\text{Prior} + \text{Likelihood} \rightarrow \text{Posterior}
$$

is the core of **Bayesian inference**.

##### <h2 style="color:#1E40AF; margin-top:0;">Why Bayesian Regression Can Be Better: Handling Real-World Constraints</h2>

<p style="margin:0 0 8px 0;">
In many real-world problems, outcome variables have <span style="color:#D97706;"><strong>natural limits</strong></span>.  
Bayesian regression allows us to encode domain knowledge as <strong>priors</strong> and ensures predictions and coefficients respect these constraints.
</p>

<hr style="border:none; height:1px; background:#CBD5E1; margin:12px 0;">

<h3 style="color:#2C5282; margin-bottom:6px;">Setup: Simple Example</h3>

<ul>
  <li>Model: \(y_i = \beta_0 + \beta_1 x_i + \varepsilon_i,\; \varepsilon_i\sim N(0,\sigma^2)\)</li>
  <li>Domain knowledge: slope \(\beta_1\) must satisfy <span style="color:#16A34A;"><strong>0 ≤ β₁ ≤ c</strong></span> for some known constant \(c\) (e.g., physical law or operational limit).</li>
  <li>Observed data (small, noisy):</li>
</ul>

<pre style="background:#FFFFFF; padding:10px; border-radius:6px; border:1px solid #E2E8F0;">
x = [1, 2, 3]
y = [2.1, 1.9, 1.8]  # small sample, noise can produce negative slope
</pre>

<p style="margin-top:8px;">
OLS gives slope estimate \(\hat\beta_1 \approx -0.45\), which violates the known constraint 0 ≤ β₁ ≤ c.
</p>

<hr style="border:none; height:1px; background:#CBD5E1; margin:12px 0;">

<h3 style="color:#2C5282; margin-bottom:6px;">1) Frequentist / OLS Limitation</h3>

<ul>
  <li>Minimizes squared error without considering domain rules.</li>
  <li>Parameters are unbounded → noisy or small data can produce impossible estimates.</li>
  <li>Cannot enforce constraints naturally; requires post hoc adjustments or constrained optimization.</li>
</ul>

<div style="background:#FEF3C7; border-left:4px solid #D97706; padding:10px; border-radius:6px; margin:8px 0;">
<strong>Consequence:</strong> OLS produces \(\hat\beta_1\approx -0.45\) (negative slope) or predictions outside realistic bounds. Not probabilistically coherent with prior knowledge.
</div>

<hr style="border:none; height:1px; background:#CBD5E1; margin:12px 0;">

<h3 style="color:#2C5282; margin-bottom:6px;">2) Bayesian Approach: Encode Constraints as Priors</h3>

<p style="margin:0 0 8px 0;">
Bayesian methods treat parameters as random variables. Priors encode your knowledge before seeing the data.
</p>

<ul>
  <li><strong>Hard constraint example:</strong> Uniform prior on [0,c]:
  $$
  \beta_1 \sim \text{Uniform}(0,c)
  $$
  Assigns zero probability to impossible slopes.</li>

  <li><strong>Soft constraint example:</strong> Truncated normal prior:
  $$
  \beta_1 \sim \text{Normal}(\mu_0, \tau^2) \text{ truncated to } [0,c]
  $$
  Centers belief near \(\mu_0\) while forbidding impossible values.</li>
</ul>

<p style="margin:6px 0;">
Bayes' rule gives the posterior:
$$
p(\beta_0,\beta_1\mid y,x) \propto p(y\mid\beta_0,\beta_1,x)\;p(\beta_0)\;p(\beta_1)
$$
Because \(p(\beta_1)=0\) outside [0,c], the posterior automatically respects constraints.
</p>

<div style="background:#ECFDF5; border-left:4px solid #16A34A; padding:10px; border-radius:6px; margin:8px 0;">
<strong>Effect:</strong> Posterior places all probability density for \(\beta_1\) inside [0,c]. Posterior mean or median reflects both data and domain knowledge. Predictions stay realistic (e.g., sales ≥ 0 and ≤ inventory limits).
</div>

<hr style="border:none; height:1px; background:#CBD5E1; margin:12px 0;">

<h3 style="color:#2C5282; margin-bottom:6px;">3) Why This Matters</h3>

<ol>
  <li><strong>Respects domain constraints:</strong> impossible values excluded by construction.</li>
  <li><strong>Combines weak data & strong knowledge:</strong> avoids absurd estimates with noisy or small datasets.</li>
  <li><strong>Probabilistic:</strong> full posterior allows uncertainty quantification.</li>
  <li><strong>Flexible:</strong> choose hard (uniform) or soft (truncated normal) priors.</li>
</ol>

<hr style="border:none; height:1px; background:#CBD5E1; margin:12px 0;">

<h3 style="color:#2C5282; margin-bottom:6px;">4) Practical Notes</h3>

<ul>
  <li>Posterior may not have closed form; use MCMC (e.g., HMC, Metropolis) or grid methods.</li>
  <li>Report posterior mean/median, credible intervals, posterior predictive checks.</li>
  <li>With large data, likelihood dominates; with small/moderate data, the prior can guide estimates reliably.</li>
</ul>

<hr style="border:none; height:1px; background:#CBD5E1; margin:12px 0;">

<h3 style="color:#1F2937; margin-bottom:6px;">5) Short Summary</h3>

<div style="background:#EEF2FF; padding:12px; border-radius:6px; border-left:4px solid #4338CA;">
If a coefficient or prediction must satisfy known bounds, encode that as a <strong>prior</strong>. Frequentist OLS ignores constraints and may produce impossible values. Bayesian regression combines data and domain knowledge, ensuring realistic predictions while still quantifying uncertainty.
</div>

<h2 style="color:#047857; margin-top:0; text-align:center;">Conjugate Priors in Bayesian Estimation</h2>

<p>Bayesian estimation combines prior knowledge with observed data to produce a posterior distribution. 
                                                                                                      How we compute the posterior depends on whether a conjugate prior exists.</p>

<hr style="border:none; height:1px; background:#34D399; margin:12px 0;">

<h3 style="color:#065F46;">1. Conjugate Prior (Closed-Form Posterior)</h3>
<p>A <strong>conjugate prior</strong> is a prior distribution that ensures the posterior is in the same family as the prior. This allows <strong>analytical computation</strong> without simulation.</p>

<ul>
  <li>Example 1: Coin Toss
    <ul>
      <li>Likelihood: \(X \sim \text{Binomial}(n, \theta)\)</li>
      <li>Prior: \(\theta \sim \text{Beta}(\alpha, \beta)\)</li>
      <li>Posterior: <span style="color:#047857;"><strong>\(\theta \mid X \sim \text{Beta}(\alpha + \text{successes}, \beta + \text{failures})\)</strong></span></li>
    </ul>
  </li>
  <li>Example 2: Normal with Known Variance
    <ul>
      <li>Likelihood: \(X_i \sim \mathcal{N}(\mu, \sigma^2)\)</li>
      <li>Prior: \(\mu \sim \mathcal{N}(\mu_0, \tau^2)\)</li>
      <li>Posterior: <span style="color:#047857;"><strong>
        \(\mu \mid X \sim \mathcal{N}\Bigg(
        \frac{\frac{\mu_0}{\tau^2} + \frac{\sum X_i}{\sigma^2}}{\frac{1}{\tau^2} + \frac{n}{\sigma^2}}, \
        \frac{1}{\frac{1}{\tau^2} + \frac{n}{\sigma^2}}
        \Bigg)\)
      </strong></span></li>
    </ul>
  </li>
</ul>

<hr style="border:none; height:1px; background:#34D399; margin:12px 0;">

<h3 style="color:#065F46;">2. Non-Conjugate / Complex Likelihood</h3>
<ul>
  <li>Likelihood is <span style="color:#B91C1C;"><strong>complex or non-standard</strong></span></li>
  <li>Prior is <span style="color:#B91C1C;"><strong>not conjugate</strong></span> to the likelihood</li>
  <li>Solution: Use <span style="color:#047857;"><strong>simulation-based methods</strong></span>
    <ul>
      <li>MCMC (Markov Chain Monte Carlo)</li>
      <li>Metropolis-Hastings / Gibbs Sampling</li>
    </ul>
  </li>
</ul>

<hr style="border:none; height:1px; background:#34D399; margin:12px 0;">

<div style="border-left:4px solid tomato; background-color:#FFFFFF; padding:10px; border-radius:6px;">
<span style="color:#065F46;"><strong>Key Point:</strong></span> Conjugacy makes Bayesian updates <span style="color:#047857;"><strong>easy and analytical</strong></span>, while non-conjugacy requires <span style="color:#B91C1C;"><strong>simulation or approximation</strong></span>.
</div>

</div>

##### To understand How and Why we do simulation , first understand below example of Integeration 
##### Whose Integeration is not possible directly 

<div style="border:2px solid #2563EB; padding:20px; border-radius:12px; background:#EFF6FF; font-family:-apple-system, BlinkMacSystemFont, 'Segoe UI', Roboto, 'Helvetica Neue', Arial;">

<h2 style="color:#1D4ED8; margin-top:0;">Monte Carlo Simulation — Evaluating Integrals Step by Step</h2>

<p>
Monte Carlo simulation approximates difficult integrals or expected values by <strong>random sampling</strong>. Instead of solving analytically, we simulate random points and compute averages.
</p>

<hr style="border:none; height:1px; background:#93C5FD; margin:12px 0;">

<h3 style="color:#1E40AF; margin-bottom:6px;">Step 1: Define the integral</h3>

<p>Consider:</p>

<div style="background:#DBEAFE; padding:10px; border-radius:6px; margin:8px 0; border-left:4px solid #2563EB;">
<p style="text-align:center; font-size:1.1em;">\( I = \int_0^1 e^{-x^2} \, dx \)</p>
<p>Analytical solution is difficult, so simulation helps.</p>
</div>

<hr style="border:none; height:1px; background:#93C5FD; margin:12px 0;">

<h3 style="color:#1E40AF; margin-bottom:6px;">Step 2: Sample random points</h3>

<div style="background:#FEF3C7; padding:10px; border-radius:6px; margin:8px 0; border-left:4px solid #F59E0B;">
<p>Generate \(N\) points uniformly from [0,1]:</p>
<p style="text-align:center; font-size:1.1em;">\( x_1, x_2, ..., x_N \sim \text{Uniform}(0,1) \)</p>
</div>

<p>These points are the "random trials" for the integral.</p>

<hr style="border:none; height:1px; background:#93C5FD; margin:12px 0;">

<h3 style="color:#1E40AF; margin-bottom:6px;">Step 3: Evaluate the function</h3>

<div style="background:#DBEAFE; padding:10px; border-radius:6px; margin:8px 0; border-left:4px solid #2563EB;">
<p style="text-align:center; font-size:1.1em;">\( f(x_1), f(x_2), ..., f(x_N), \quad f(x_i) = e^{-x_i^2} \)</p>
<p>Each \(f(x_i)\) is like one "measurement" of the integral under a randomly chosen x.</p>
</div>

<hr style="border:none; height:1px; background:#93C5FD; margin:12px 0;">

<h3 style="color:#1E40AF; margin-bottom:6px;">Step 4: Approximate the integral</h3>

<div style="background:#FEF3C7; padding:10px; border-radius:6px; margin:8px 0; border-left:4px solid #F59E0B;">
<p style="text-align:center; font-size:1.1em;">\( I \approx \frac{1}{N} \sum_{i=1}^{N} f(x_i) \)</p>
<p>- Larger N → more accurate approximation.</p>
</div>

<hr style="border:none; height:1px; background:#93C5FD; margin:12px 0;">

<h3 style="color:#1E40AF; margin-bottom:6px;">Step 5: Numerical example</h3>

<div style="background:#DBEAFE; padding:10px; border-radius:6px; margin:8px 0; border-left:4px solid #2563EB;">
<ul>
  <li>Take \(N=5\) points: \( x = [0.1, 0.3, 0.5, 0.7, 0.9] \)</li>
  <li>Evaluate function: \( f(x) = e^{-x^2} \Rightarrow [0.990, 0.913, 0.779, 0.612, 0.419] \)</li>
  <li>Approximate integral:</li>
</ul>
<p style="text-align:center; font-size:1.1em;">\( I \approx \frac{0.990 + 0.913 + 0.779 + 0.612 + 0.419}{5} = 0.743 \)</p>
<p>Close to the true value \( \sim 0.7468 \) even with only 5 samples.</p>
</div>

<hr style="border:none; height:1px; background:#93C5FD; margin:12px 0;">

<h3 style="color:#1E40AF; margin-bottom:6px;">Step 6: Connection to Bayesian estimation</h3>

<div style="background:#FEF3C7; padding:10px; border-radius:6px; margin:8px 0; border-left:4px solid #F59E0B;">
<ul>
  <li>Posterior expectations require integration over parameter space:</li>
  <p style="text-align:center; font-size:1.1em;">\( p(\theta | y) \propto p(y | \theta) \, p(\theta) \)</p>
  <li>High dimensions make direct integration infeasible.</li>
  <li>MCMC or other simulations approximate the posterior:</li>
  <p style="text-align:center; font-size:1.1em;">\( \text{posterior mean} \approx \frac{1}{N} \sum_{i=1}^{N} \theta_i \)</p>
  <li>Thus, simulation provides a practical way to compute otherwise intractable posterior quantities.</li>
</ul>
</div>

<h2 style="color:#1D4ED8; margin-top:0;">Metropolis-Hastings (MH) Algorithm Explained</h2>

<p>
The Metropolis-Hastings algorithm is a <strong>Markov Chain Monte Carlo (MCMC)</strong> method used to generate samples from a <em>complex probability distribution</em> when direct sampling is difficult. It produces a sequence of samples that, over time, follows the target distribution.
</p>

<hr style="border:none; height:1px; background:#93C5FD; margin:12px 0;">

<h3 style="color:#1E40AF;">Step 1: Problem Setup</h3>
<ul>
  <li>Target distribution: <strong>p(x) = e^{-x²}, x ≥ 0</strong></li>
  <li>Goal: generate samples x₁, x₂, ..., x_N from p(x)</li>
  <li>Direct sampling is difficult → use MH algorithm</li>
</ul>

<hr style="border:none; height:1px; background:#93C5FD; margin:12px 0;">

<h3 style="color:#1E40AF;">Step 2: MH Algorithm Steps</h3>
<ol>
  <li><strong>Initialize:</strong> choose a starting point x₀</li>
  <li><strong>Propose:</strong> generate candidate x* from a proposal distribution q(x* | x_t), e.g., Normal(x_t, σ²)</li>
  <li><strong>Compute acceptance probability:</strong>
    <p style="text-align:center;">
    \( \alpha = \min\Big(1, \frac{p(x^*) \, q(x_t | x^*)}{p(x_t) \, q(x^* | x_t)}\Big) \)
    </p>
    <p>For symmetric proposals (Normal), q terms cancel:</p>
    <p style="text-align:center;">
    \( \alpha = \min(1, \frac{p(x^*)}{p(x_t)}) \)
    </p>
  </li>
  <li><strong>Accept/Reject:</strong> draw u ~ Uniform(0,1)
    <ul>
      <li>If u ≤ α → accept: x_{t+1} = x*</li>
      <li>Else → reject: x_{t+1} = x_t</li>
    </ul>
  </li>
  <li><strong>Repeat:</strong> repeat steps 2–4 for N iterations to generate samples x₁, x₂, ..., x_N</li>
</ol>

<hr style="border:none; height:1px; background:#93C5FD; margin:12px 0;">

<h3 style="color:#1E40AF;">Step 3: Understanding Acceptance Probability</h3>
<p>MH uses <strong style="color:#16A34A;">α</strong> to balance exploration and correctness:</p>

<table style="width:100%; border:1px solid #CBD5E1; border-collapse:collapse; margin-top:6px;">
<tr style="background:#DBEAFE;"><th style="border:1px solid #CBD5E1; padding:4px;">x</th><th style="border:1px solid #CBD5E1; padding:4px;">p(x)</th><th style="border:1px solid #CBD5E1; padding:4px;">Intuition</th></tr>
<tr><td style="border:1px solid #CBD5E1; padding:4px;">0</td><td style="border:1px solid #CBD5E1; padding:4px;">1.0</td><td>Very likely (peak)</td></tr>
<tr><td style="border:1px solid #CBD5E1; padding:4px;">1</td><td style="border:1px solid #CBD5E1; padding:4px;">0.3679</td><td>Moderately likely</td></tr>
<tr><td style="border:1px solid #CBD5E1; padding:4px;">2</td><td style="border:1px solid #CBD5E1; padding:4px;">0.0183</td><td>Very unlikely (tail)</td></tr>
</table>

<p style="text-align:center;">
\(\alpha = \min(1, p(x^*)/p(x_t))\)
</p>
<ul>
  <li>High-probability x* → α ≈ 1 → almost always accepted</li>
  <li>Low-probability x* → α < 1 → sometimes rejected</li>
  <li>This allows the chain to explore the full distribution and not get stuck at a peak.</li>
</ul>

<hr style="border:none; height:1px; background:#93C5FD; margin:12px 0;">

<h3 style="color:#1E40AF;">Step 4: Concrete Numeric Example</h3>
<ul>
  <li>Target: p(x) = e^{-x²}, x ≥ 0</li>
  <li>Current position: <strong style="color:#2563EB;">x_t = 1.0</strong></li>
  <li>Proposed point: <strong style="color:#2563EB;">x* = 1.5</strong></li>
  <li>Compute probabilities:
    <p style="text-align:center;">
    p(x_t) = e^{-1^2} ≈ 0.3679 <br>
    p(x^*) = e^{-1.5^2} ≈ 0.1054
    </p>
  </li>
  <li>Acceptance probability:
    <p style="text-align:center;">
    \(\alpha = \min(1, 0.1054 / 0.3679) \approx 0.286\)
    </p>
  </li>
  <li>Draw u ~ Uniform(0,1), e.g., u = 0.2 → u ≤ α → accept x* = 1.5</li>
  <li>If u = 0.5 → u > α → reject → stay at x_t = 1.0</li>
</ul>

<hr style="border:none; height:1px; background:#93C5FD; margin:12px 0;">

<h3 style="color:#1E40AF;">Step 5: Summary Table</h3>
<table style="width:100%; border:1px solid #CBD5E1; border-collapse:collapse; margin-top:6px;">
<tr style="background:#DBEAFE;">
<th style="border:1px solid #CBD5E1; padding:4px;">Step</th>
<th style="border:1px solid #CBD5E1; padding:4px;">Math / Formula</th>
<th style="border:1px solid #CBD5E1; padding:4px;">Intuition</th>
</tr>
<tr><td>Propose x*</td><td>x* ~ q(x* | x_t)</td><td>Pick a new candidate near current point</td></tr>
<tr><td>Compute α</td><td>\(\alpha = \min(1, p(x^*)/p(x_t))\)</td><td>Decide acceptance probability based on likelihood</td></tr>
<tr><td>Draw u</td><td>u ~ Uniform(0,1)</td><td>Random number to compare with α</td></tr>
<tr><td>Accept/Reject</td><td>if u ≤ α → x_{t+1}=x*, else x_{t+1}=x_t</td><td>Move or stay → explores distribution correctly</td></tr>
</table>

<hr style="border:none; height:1px; background:#93C5FD; margin:12px 0;">

<h3 style="color:#1E40AF;">Step 6: Intuition</h3>
<ul>
  <li>High probability points → almost always accepted</li>
  <li>Low probability points → sometimes accepted → allows exploration</li>
  <li>Over many iterations, the chain spends time at x proportionally to p(x) → correct sampling</li>
  <li>MH effectively transforms a <strong>difficult sampling problem</strong> into repeated <strong>simple accept/reject decisions</strong></li>
</ul>

</div>

```python
import numpy as np
import matplotlib.pyplot as plt

def p(x):
    return np.exp(-x**2)

N = 5000
samples = []
x = 0.0
sigma = 0.5

for _ in range(N):
    x_new = x + np.random.normal(0, sigma)
    if x_new < 0: x_new = -x_new
    alpha = min(1, p(x_new)/p(x))
    if np.random.rand() < alpha:
        x = x_new
    samples.append(x)

plt.hist(samples, bins=50, density=True, alpha=0.7, color="#3B82F6")
x_vals = np.linspace(0, 3, 200)
plt.plot(x_vals, p(x_vals)/sum(p(x_vals)*0.01), 'r', lw=2)
plt.title("Metropolis-Hastings Sampling from p(x) = exp(-x²)")
plt.show()
```

```python
import pandas as pd
import numpy as np
####
df=pd.read_csv('ad_data.csv')
df=df[['Sales','TV', 'Digital', 'Radio']]
df.head(2)
```

```python
import statsmodels.api as sm

##### Define predictor and target
X = df[["TV", "Digital", "Radio"]]
X = sm.add_constant(X)  #### add intercept
y = df["Sales"]

##### Fit OLS
ols_model = sm.OLS(y, X).fit()

##### Summary
print(ols_model.summary())
```

```python
import numpy as np
from scipy.stats import norm, truncnorm

########
### Step 1 : Define Functions
####################################################

### this function will draw samples from a truncated normal distribution
### mean: center of normal, sd: standard deviation, lower: minimum allowed, upper: maximum allowed
### size: number of samples
def truncated_normal_prior(mean, sd, lower, upper, size=1):
    a, b = (lower - mean)/sd, (upper - mean)/sd
    samples = truncnorm.rvs(a, b, loc=mean, scale=sd, size=size)
    return samples

########

### this function will draw samples from a normal distribution
### mean: center, sd: standard deviation, size: number of samples
def normal_prior(mean, sd, size=1):
    samples = norm.rvs(loc=mean, scale=sd, size=size)
    return samples

########

### this function computes likelihood of observed sales given predicted mean and sigma
### sales_obs: observed y values
### mu: predicted y values
### sigma: standard deviation of noise
def likelihood(sales_obs, mu, sigma):
    likelihood_vals = norm.pdf(sales_obs, loc=mu, scale=sigma)
    return likelihood_vals

########

### this function computes the unnormalized posterior probability
### beta_0, beta_tv, beta_digital, beta_radio: coefficients
### sigma: noise standard deviation
### X: numpy array of predictors
### y: observed sales
def posterior(beta_0, beta_tv, beta_digital, beta_radio, sigma, X, y):
    mu = beta_0 + beta_tv*X[:,0] + beta_digital*X[:,1] + beta_radio*X[:,2]
    lik_vals = norm.pdf(y, loc=mu, scale=sigma)
    post_prob = np.prod(lik_vals)
    return post_prob
```

```python
### step 2: extract predictor matrix and response
X = df[["TV", "Digital", "Radio"]].values
y = df["Sales"].values
```

```python
### step 3: monte carlo sampling for posterior
### number of posterior samples to generate
n_samples = 3000
posterior_samples = []

for i in range(n_samples):
    ### sample priors for each coefficient
    b0 = normal_prior(50, 20)
    btv = truncated_normal_prior(3, 2, 0, 10)
    bdigital = truncated_normal_prior(4, 2, 0, 10)
    bradio = truncated_normal_prior(2, 1, 0, 10)
    
    ### sample sigma from positive normal (half-normal)
    sigma = abs(normal_prior(0, 10))
    
    ### compute unnormalized posterior
    post_prob = posterior(b0[0], btv[0], bdigital[0], bradio[0], sigma[0], X, y)
    
    ### store sampled parameters
    posterior_samples.append([b0[0], btv[0], bdigital[0], bradio[0], sigma[0]])
```

```python
### step 4: convert posterior samples to dataframe
posterior_samples = np.array(posterior_samples)
posterior_df = pd.DataFrame(posterior_samples, columns=["beta_0","beta_tv","beta_digital","beta_radio","sigma"])
```

<h2 style="color:#4B0082; margin-top:0;">What if we incorporate <span style="color:#16A34A;">Adstock</span> and <span style="color:#DC2626;">Hill</span> functions also in this?</h2>

<!-- Adstock Function -->
<p style="font-size:17px; color:#4B5563;"><b>Adstock Function</b></p>

$$
\text{Adstock}_t 
= \text{Spend}_t 
+ \lambda\,\text{Spend}_{t-1}
+ \lambda^2\,\text{Spend}_{t-2}
+ \dots
$$

<p style="color:#374151;">Recursive form:</p>

$$
\text{Adstock}_t
= \text{Spend}_t + \lambda\,\text{Adstock}_{t-1}
$$

<hr>

<!-- Hill Function -->
<p style="font-size:17px; color:#4B5563;"><b>Hill Saturation Function</b></p>

$$
\text{Hill}(x)=
\frac{
x^{\color{#2563EB}{\alpha}}
}{
\theta^{\color{#2563EB}{\alpha}} + x^{\color{#2563EB}{\alpha}}
}
$$

<hr>

<!-- Final MMM -->
<p style="font-size:17px; color:#4B5563;"><b>Final MMM Equation with Adstock + Hill</b></p>

$$
\text{Sales}_t =
\beta_{\color{#7C3AED}{0}}
+ 
\beta_{\color{#16A34A}{TV}}\,\text{Hill}(\text{Adstock}_{TV,t})
+ 
\beta_{\color{#2563EB}{Digital}}\,\text{Hill}(\text{Adstock}_{Digital,t})
+ 
\beta_{\color{#DC2626}{Radio}}\,\text{Hill}(\text{Adstock}_{Radio,t})
+ \varepsilon_t
$$


<!-- Use MCMC -->
<p style="font-size:17px; color:#4B5563;"><b>Use MCMC : Metropolis–Hastings</b></p>

<li>does not need closed-form posterior</li>
<li>handles nonlinear adstock + hill</li>
<li>handles correlated parameters</li>
