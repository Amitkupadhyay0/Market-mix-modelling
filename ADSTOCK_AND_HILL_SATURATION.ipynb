<h3 style="color:#000; font-weight:bold; margin-top:0;">
What is Adstock?
</h3>

<ul style="color:#333; line-height:1.7;">

<li><b>Adstock</b> is a technique in Marketing Mix Modeling that converts raw advertising spend into a more realistic and effective measure of impact.</li>

<li>It accounts for the fact that advertising continues to influence consumers for several days, not just on the day it occurs.</li>

<li>Adstock introduces a <b>carryover effect</b> so that part of yesterday’s advertising still contributes to today’s impact.</li>

<li>A <b>decay factor</b> is applied so that recent ads have stronger influence and older ads gradually lose effect.</li>

<li>This prevents the incorrect assumption that “spend on a particular day affects only that day’s sales.”</li>

<li>Adstock helps capture three real behaviours of advertising:
    <ul>
        <li><b>Delayed impact</b> — consumers may respond after a few days.</li>
        <li><b>Memory decay</b> — impact reduces gradually over time.</li>
        <li><b>Build-up effect</b> — repeated ads across days reinforce each other.</li>
    </ul>
</li>

<li>Overall, Adstock transforms advertising spend into a smoother and more realistic signal that better represents actual consumer response.</li>

</ul>

</div>


$\small \text{Adstock}_{t} = \text{Spend}_{t} + \lambda \cdot \text{Spend}_{t-1} + \lambda^2 \cdot \text{Spend}_{t-2} + \dots$

For Example :
$\small \text{Adstock}_{3} = \text{Spend}_{3} + \lambda \cdot \text{Spend}_{2} + \lambda^2 \cdot \text{Spend}_{1}$

#### How do we compute Adstock?

Let's explore the methods.

### **<span style="color:violet">1. Exponential Distribution</span>**

**Why Exponential?**

- Models decay of advertising effect over time naturally.  

```python
import numpy as np
import matplotlib.pyplot as plt

############# Parameter ############
days = np.arange(0, 10, 1)
lambdas = [0.3, 0.5, 0.7]
colors = ['tomato', 'violet', 'orange']

plt.figure(figsize=(6,3))

for lam, color in zip(lambdas, colors):
    decay = np.exp(-lam * days)
    plt.plot(days, decay, marker='o', color=color, label=f'λ = {lam}', linewidth=2)

plt.title('Exponential Decay of Ad Effect', fontsize=10, color='tomato')
plt.xlabel('Days', fontsize=8)
plt.ylabel('Effect', fontsize=8)
plt.xticks(fontsize=8)
plt.yticks(fontsize=8)
plt.legend(loc='upper right', fontsize=8)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

![Exponential Decay Visualization](EXP_D_DRAW_1.png)

### **<span style="color:violet">2. Weibull Distribution</span>**

**Why Weibull?**

###### Weibull Decay: Shape (k) & Scale (λ) Effects

- **Shape (k):** controls the curve form  
  - k < 1 → fast initial decay  
  - k = 1 → exponential decay  
  - k > 1 → slow build-up before decay  

- **Scale (λ):** stretches or compresses the timeline  
  - Larger λ → slower decay, effect lasts longer  
  - Smaller λ → faster decay  

**Observation:**  
By adjusting k and λ, Weibull can model **instant, delayed, or prolonged ad effects**, making it more flexible than simple exponential decay.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import weibull_min

###################################
days = np.arange(0, 10, 1)

# #################################
params = [
    {'k': 1.5, 'scale': 1, 'color': 'tomato', 'label': 'k=1.5, λ=1'},
    {'k': 2.5, 'scale': 5, 'color': 'violet', 'label': 'k=2.5, λ=5'}
]

plt.figure(figsize=(6,3))

############################################
for p in params:
    decay = weibull_min.pdf(days, p['k'], scale=p['scale'])
    decay = decay / decay.max() 
    plt.plot(days, decay, marker='o', color=p['color'], label=p['label'], linewidth=2)

plt.title('Weibull Decay: Shape & Scale Effects', fontsize=10, color='tomato')
plt.xlabel('Days', fontsize=8)
plt.ylabel('Normalized Effect', fontsize=8)
plt.xticks(fontsize=8)
plt.yticks(fontsize=8)
plt.legend(loc='upper right', fontsize=8)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

### **<span style="color:green"> HOW WE ESTIMATE PARAMTERS ??**

### <span style="font-size:85%; color:#CC3366"><b>A. Estimate λ by Grid Search (considering only marketing factor)</b></span>

**TV Spend Data**

| Day | TV_Spend |
|-----|----------|
| 1   | 100      |
| 2   | 0        |
| 3   | 50       |
| 4   | 10       |

---

##### <span style="font-size:85%"><b>1. Compute exponential adstock for each candidate λ</b></span>

The adstock formula:

$$
\small
\text{Adstock}_{t}(\lambda)
=
\text{Spend}_{t} + \lambda \cdot \text{Adstock}_{t-1}
$$

---

##### <span style="font-size:85%"><b>2. Fit simple regression for each λ</b></span>

$$
\small
\text{Sales}_t
=
\beta_0(\lambda)
+
\beta_1(\lambda)\,\text{Adstock}_{t}(\lambda)
+
\varepsilon_t
$$

---

##### <span style="font-size:85%"><b>3. Evaluate model fit & pick best λ</b></span>

| λ   | R²   | RMSE |
|-----|------|------|
| 0.2 | 0.52 | 14.2 |
| 0.4 | 0.76 | 8.1  |
| 0.6 | 0.68 | 10.5 |

**Best λ = 0.4**

---

##### <span style="font-size:85%; color:#BB2222"><b>Limitation</b></span>

Grid search ignores:

- Other marketing channels  
- Non-marketing factors (seasonality, holidays, price)

This can bias λ.

---

### <span style="font-size:90%; color:#3366CC"><b>B. Estimate λ directly inside the full MMM</b></span>

Instead of computing adstock separately, the MMM **learns λ automatically**.

---

### <span style="font-size:85%"><b>Full Multi-Channel MMM Equation</b></span>

$$
\small
\text{Sales}_t
=
\underbrace{\beta_0}_{\text{baseline}}
+
\sum_{c\in\mathcal{C}} \underbrace{\beta_c \, \text{Adstock}_{c,t}(\lambda_c)}_{\text{marketing factor}}
+
\underbrace{\sum_j \gamma_j X_{j,t}}_{\text{non-marketing factors}}
+
\varepsilon_t
$$

Where the channel-specific adstock is:

$$
\small
\text{Adstock}_{c,t}(\lambda_c)
=
\text{Spend}_{c,t} + \lambda_c \cdot \text{Adstock}_{c,t-1},
\qquad
0 \le \lambda_c < 1
$$

---
##### <span style="font-size:85%; color:#008866"><b>Why joint estimation is better?</b></span>

- λ is estimated **with** other coefficients 
- Other drivers (price, promos, holidays) included

<h2 style="
    color:#d35400;
    font-weight:bold;
    text-align:center;
">Hill Saturation: Modeling Diminishing Returns</h2>

<p style="color:#333; font-size:15px;">
Even after accounting for adstock (carryover), more spend 
<b style="color:#c0392b;">does not always give linear sales</b>.  
The Hill function models this <b style="color:#8e44ad;">saturation effect</b>.
</p>

<hr style="border:1px solid #ff7f50;">

<h3 style="color:#d35400;">Hill Function Formula</h3>

<p style="color:#2c3e50;">
\[
\small
\text{Effect}_t = \frac{\text{Adstock}_t^n}{\text{Adstock}_t^n + \alpha^n}
\]
</p>

<ul style="color:#2c3e50; font-size:15px;">
  <li><b>Adstockₜ</b> → carryover-adjusted spend</li>
  <li><b>α</b> → spend at 50% maximum effect</li>
  <li><b>n</b> → curve steepness</li>
</ul>

<hr style="border:1px solid #ff7f50;">

<h3 style="color:#d35400;">Simple Example</h3>

<p style="color:#2c3e50;">Adstocked TV spend for 4 days:</p>

<p style="color:#2c3e50;">
Hill effect for Day 3:
</p>

<p style="color:#2c3e50;">
\[
\text{Effect}_3 = \frac{75^2}{75^2 + 50^2} \approx 0.69
\]
</p>

<p style="color:#c0392b; font-weight:bold;">
This means <b>69% of max possible effect</b> is achieved for Day 3 spend.
</p>

<hr style="border:1px solid #ff7f50;">

<h3 style="color:#d35400;">Key Points</h3>

<ul style="color:#2c3e50; font-size:15px;">
  <li>Converts <b style="color:#16a085;">adstocked spend → effective sales contribution</b></li>
  <li>Captures <b style="color:#8e44ad;">non-linear, diminishing returns</b></li>
  <li>Works for any channel</li>
</ul>

</div>

```python
import numpy as np
import matplotlib.pyplot as plt
################################
alpha = 50
n = 2
###############################
x = np.linspace(0, 200, 500)
hill = (x**n) / (x**n + alpha**n)
plt.figure(figsize=(6,4))
plt.plot(x, hill, linewidth=3)
plt.axvline(alpha, linestyle="--", linewidth=2)
plt.scatter([alpha], [(alpha**n)/(alpha**n + alpha**n)], s=80, zorder=5)
plt.text(alpha+1, 0.05, "α (50% saturation)", fontsize=12)
plt.text(10, 0.9, "Hill Saturation Curve", fontsize=6, fontweight="bold")
plt.xlabel("Adstocked Spend", fontsize=6)
plt.ylabel("Saturation Effect", fontsize=6)
plt.title("Hill Function: Diminishing Returns", fontsize=8, fontweight="bold")
plt.grid(alpha=0.3)
plt.ylim(0,1.05)
plt.show()
```

<h2 style="
    color:#4b0082;
    text-align:center;
    font-weight:bold;
">
Comprehensive MMM Model with Adstock + Hill
</h2>

<p style="color:#333;">
Modern Marketing Mix Modeling (MMM) uses:<br>
<b style="color:#8e44ad;">Adstock</b> for carryover effects, and  
<b style="color:#c0392b;">Hill function</b> for saturation / diminishing returns.  
Below is the unified full equation used in Bayesian and Robyn-style MMM.
</p>

<hr style="border:1px solid #6a5acd;">

<h3 style="color:#4b0082;">Full Sales Response Equation</h3>

<p style="color:#2c3e50;">
\[
\text{Sales}_t
=
\underbrace{\beta_0}_{\text{baseline}}
+
\sum_{c\in\mathcal{C}}
\underbrace{
\beta_c \cdot 
H\!\left(\text{Adstock}_{c,t}(\lambda_c),\, \alpha_c,\, n_c\right)
}_{\text{marketing\ impact}}
+
\underbrace{\sum_j \gamma_j X_{j,t}}_{\text{control\ variables}}
+
\varepsilon_t
\]
</p>

<p style="color:#2c3e50;">
Where each channel has parameters  
\(\beta_c, \lambda_c, \alpha_c, n_c\) to be estimated.
</p>

<div style="
    background:#ffffff;
    border-left:5px solid #4b0082;
    padding:12px;
    margin:12px 0;
">

<h3 style="color:#4b0082; margin-top:0;">1. Adstock (Carryover) Function</h3>

<p style="color:#2c3e50;">
\[
\text{Adstock}_{c,t}(\lambda_c)
=
\text{Spend}_{c,t}
+
\lambda_c \cdot \text{Adstock}_{c,t-1},
\qquad
0 \le \lambda_c < 1
\]
</p>

<ul style="color:#2c3e50;">
  <li><b>λ₍c₎</b>: decay/carryover rate (estimated)</li>
  <li>Captures memory effect of media</li>
</ul>

</div>

<div style="
    background:#ffffff;
    border-left:5px solid #c0392b;
    padding:12px;
    margin:12px 0;
">

<h3 style="color:#c0392b; margin-top:0;">2. Hill Saturation Function</h3>

<p style="color:#2c3e50;">
\[
H(A_{c,t}; \alpha_c, n_c)
=
\frac{A_{c,t}^{n_c}}
{A_{c,t}^{n_c} + \alpha_c^{\,n_c}}
\]
</p>

<ul style="color:#2c3e50;">
  <li><b>A₍c,t₎</b>: adstocked spend</li>
  <li><b>α₍c₎</b>: spend at 50% effect (estimated)</li>
  <li><b>n₍c₎</b>: slope/steepness (estimated)</li>
</ul>

</div>

<hr style="border:1px solid #6a5acd;">

<h3 style="color:#4b0082;">Combined Channel Contribution</h3>

<p style="color:#2c3e50;">
\[
\text{Effect}_{c,t}
=
\beta_c \cdot
\frac{
\left(\text{Adstock}_{c,t}(\lambda_c)\right)^{n_c}
}{
\left(\text{Adstock}_{c,t}(\lambda_c)\right)^{n_c}
+
\alpha_c^{\,n_c}
}
\]
</p>

<p style="color:#2c3e50;">
This transforms:
<br>
Spend → Adstock → Saturation → Final Sales Impact.
</p>

<hr style="border:1px solid #6a5acd;">

<h3 style="color:#4b0082;">Alternative Functional Form (Equivalent)</h3>

<p style="color:#2c3e50;">
\[
\text{Sales}_t =
\beta_0 +
\sum_{c}
f_c\!\left(
\text{Spend}_{c,t} \mid 
\theta_c = (\lambda_c, \alpha_c, n_c, \beta_c)
\right)
+
\sum_j \gamma_j X_{j,t}
+
\varepsilon_t
\]
</p>

<p style="color:#2c3e50;">
Where  
\[
f_c = 
\beta_c \cdot
H\!\left(
\text{Adstock}(\text{Spend},\lambda_c),\,
\alpha_c,\,
n_c
\right)
\]
</p>

<p style="color:#2c3e50;">
This makes it explicit that  
\(\theta_c = (\lambda_c, \alpha_c, n_c, \beta_c)\)  
are all parameters learned during modeling.
</p>
